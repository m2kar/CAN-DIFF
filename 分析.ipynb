```python
import json
import os
```

```python
with open("base.json",encoding="utf8") as fp:
    base=json.load(fp)["ld"]
```

```python
with open("a1.json",encoding="utf8") as fp:
    a1=json.load(fp)["ld"]
```

```python
def get_idlist(ll):
    idlist={}
    for each in ll:
        if each.get("type","")!="comment":
            idlist[each["id"]]=idlist.get(each["id"],0)+1
    return idlist
```

```python
def readlistfromjson(filename):
    with open(filename,encoding="utf8") as fp:
        ll=json.load(fp)["ld"]
    return ll
```

```python
base_idlist=get_idlist(base)
```

```python
a1=readlistfromjson("a1.json")
a2=readlistfromjson("a2.json")
a3=readlistfromjson("a3.json")
a1_idlist=get_idlist(a1)
a2_idlist=get_idlist(a2)
a3_idlist=get_idlist(a3)
```

```python
def json2csv(filename):
    with open(filename,encoding="utf8") as jsonfp:
        ll=json.load(jsonfp)["ld"]
    with open(os.path.splitext(filename)[0]+".csv","w",encoding="utf8") as csvfp:
        for each in ll:
            if each.get("type","")!="comment":
                csvfp.write("{0},{1},".format(each["timestamp"],each["id"]))
                csvfp.write(",".join([hex(i) for i in each["data"]]))
                csvfp.write("\n")
#                 idlist[each["id"]]=idlist.get(each["id"],0)+1
```

```python
json2csv("a1.json")
json2csv("base.json")
```
